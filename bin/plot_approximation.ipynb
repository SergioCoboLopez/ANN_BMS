```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
import matplotlib.gridspec as gridspec
import ast
import sys
sys.path.append('machine-scientist/')
sys.path.append('machine-scientist/Prior/')
from mcmc import *
from parallel import *
from fit_prior import read_prior_par
from sklearn.metrics import mean_squared_error
from sklearn.metrics import root_mean_squared_error
from sklearn.metrics import mean_absolute_error
from scipy.stats import bootstrap
```

```python
#Read error dataframe for all simulations
function='tanh'
errors=pd.read_csv('../data/'+ 'errors_' + str(function) + '.csv',index_col=0)
display(errors)
```

```python
columns=["sigma","rmse_nn_train", "rmse_nn_test", "rmse_mdl_train", "rmse_mdl_test"]
errors_test=errors[columns]
display(errors_test)

errors_test=pd.melt(errors_test, id_vars=["sigma", "rmse_nn_test", "rmse_mdl_test"], var_name="error_train", value_name= "value_train")
display(errors_test)

errors_test=pd.melt(errors_test, id_vars=["sigma", "error_train", "value_train"], var_name="error_test", value_name= "value_test")
display(errors_test)
```

```python
sns.catplot(data=errors_test, x='sigma', y='value_train', hue="error_train", kind='point', estimator='median',errorbar=('ci', 95),n_boot=100000, 
            palette=['red', 'green'])
plt.yscale("log")

sns.catplot(data=errors_test, x='sigma', y='value_test', hue="error_test", kind='point', estimator='median',errorbar=('ci', 95),n_boot=100000, 
            palette=['red', 'green'])
```

```python
sns.catplot(data=errors, x='sigma', y='rmse_mdl_train')
```

```python
all_errors=pd.read_csv('../data/'+ 'all_errors_' + '.csv',index_col=0)
display(all_errors)

columns=["sigma", "function","rmse_nn_train", "rmse_nn_test", "rmse_mdl_train", "rmse_mdl_test"]
all_errors=all_errors[columns]
display(all_errors)
```

```python
all_errors=pd.melt(all_errors,id_vars=["sigma","function","rmse_nn_test","rmse_mdl_test"], var_name="error_train",value_name= "value_train")
all_errors=pd.melt(all_errors,id_vars=["sigma","function", "error_train", "value_train"], var_name="error_test", value_name= "value_test")

display(all_errors)
```

```python
errors_tanh=all_errors[all_errors['function']=='tanh']
errors_leaky=all_errors[all_errors['function']=='leaky_ReLU']

display(errors_tanh)
display(errors_leaky)
```

```python
#Plot results
#Define figure size in cm                                                                           
cm = 1/2.54 #convert inch to cm                                                                     
width = 8*cm; height=6*cm

Extensions=['.png', '.pdf']

#Fonts and sizes                                                                                    
size_axis=12;size_ticks=10;size_title=5
line_w=1;marker_s=3 #width and marker size                                                          
m_size=6

output_path='../results/'

resamples=100000; bs_seed=1111
```

```python
#tanh - rmse train
fig=figure(figsize=(width,height), dpi=300)
sns.catplot(data=errors_tanh, x='sigma', y='value_train', hue="error_train", kind='point', estimator='median',
            errorbar=('ci', 95),n_boot=resamples, seed=bs_seed, palette=['red', 'green'],legend_out=False)
plt.yscale("log")

#labels, limits, and ticks
plt.xlabel(r'$\sigma$',fontsize=size_axis) 
plt.ylabel('rmse train',fontsize=size_axis)
plt.xticks(fontsize=size_ticks);plt.yticks(fontsize=size_ticks)
#plt.legend(loc='lower right',fontsize=size_ticks,frameon=False)

plt.ylim(2e-4,0.6)


name_fig='tanh_rmse_train.svg'
plt.savefig(output_path+name_fig,dpi=300)
```

```python
#tanh - rmse test
fig=figure(figsize=(width,height), dpi=300)

sns.catplot(data=errors_tanh, x='sigma', y='value_test', hue="error_test", kind='point', estimator='median',
            errorbar=('ci', 95),n_boot=resamples, seed=bs_seed, palette=['red', 'green'],legend_out=False)
plt.yscale("log")

#labels, limits, and ticks
plt.xlabel(r'$\sigma$',fontsize=size_axis) 
plt.ylabel('rmse train',fontsize=size_axis)
#plt.legend(loc='lower right',fontsize=size_ticks,frameon=False)

plt.ylim(2e-4,0.6)
name_fig='tanh_rmse_test.svg'
plt.savefig(output_path+name_fig,dpi=300)
```

```python
#leaky - rmse train
fig=figure(figsize=(width,height), dpi=300)

sns.catplot(data=errors_leaky, x='sigma', y='value_train', hue="error_train", kind='point', estimator='median',
            errorbar=('ci', 95),n_boot=resamples, seed=bs_seed, palette=['red', 'green'],legend_out=False)
plt.yscale("log")

#labels, limits, and ticks
plt.xlabel(r'$\sigma$',fontsize=size_axis) 
plt.ylabel('rmse train',fontsize=size_axis)
#plt.legend(loc='lower right',fontsize=size_ticks,frameon=False)

plt.ylim(4e-4,0.6)
name_fig='leaky_rmse_train.svg'
plt.savefig(output_path+name_fig,dpi=300)
```

```python
#leaky - rmse test
fig=figure(figsize=(width,height), dpi=300)

sns.catplot(data=errors_leaky, x='sigma', y='value_test', hue="error_test", kind='point', estimator='median',
            errorbar=('ci', 95),n_boot=resamples, seed=bs_seed, palette=['red', 'green'],legend_out=False)
plt.yscale("log")

#labels, limits, and ticks
plt.xlabel(r'$\sigma$',fontsize=size_axis) 
plt.ylabel('rmse train',fontsize=size_axis)
#plt.legend(loc='lower right',fontsize=size_ticks,frameon=False)

plt.ylim(4e-4,0.6)
name_fig='leaky_rmse_test.svg'
plt.savefig(output_path+name_fig,dpi=300)
```
